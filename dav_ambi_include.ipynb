```python
annotators_en = english.annotators.to_numpy().flatten().tolist()
annotators_es = spanish.annotators.to_numpy().flatten().tolist()
```

```python
Dict_CW = {}
sexist_en = sexist_en.set_index(sexist_en['id_EXIST'])
non_sexist_en = non_sexist_en.set_index(non_sexist_en['id_EXIST'])
for (i, ans) in enumerate(annotators_en):
	index = str(200000 + i + 1)
	if index in sexist_en.index:
		for (j, individual) in enumerate(ans):
			if Dict_CW.get(individual) is not None:
				if sexist_en.loc[sexist_en.id_EXIST==index]['labels_task1'].tolist()[0][j] == 'YES':
					Dict_CW[individual][0] = Dict_CW[individual][0] + 1
				else:
					Dict_CW[individual][1] = Dict_CW[individual][1] + 1
			else:
				if sexist_en.loc[sexist_en.id_EXIST==index]['labels_task1'].tolist()[0][j] == 'YES':
					Dict_CW[individual] = [1, 0]
				else:
					Dict_CW[individual] = [0, 1]
	elif index in non_sexist_en.index:
		for (j, individual) in enumerate(ans):
			if Dict_CW.get(individual) is not None:
				if non_sexist_en.loc[non_sexist_en.id_EXIST==index]['labels_task1'].tolist()[0][j] == 'NO':
					Dict_CW[individual][0] = Dict_CW[individual][0] + 1
				else:
					Dict_CW[individual][1] = Dict_CW[individual][1] + 1
			else:
				if non_sexist_en.loc[non_sexist_en.id_EXIST==index]['labels_task1'].tolist()[0][j] == 'NO':
					Dict_CW[individual] = [1, 0]
				else:
					Dict_CW[individual] = [0, 1]
```

```python
sexist_es = sexist_es.set_index(sexist_es['id_EXIST'])
non_sexist_es = non_sexist_es.set_index(non_sexist_es['id_EXIST'])
for (i, ans) in enumerate(annotators_es):
	index = str(100000 + i + 1)
	if index in sexist_es.index:
		for (j, individual) in enumerate(ans):
			if Dict_CW.get(individual) is not None:
				if sexist_es.loc[sexist_es.id_EXIST==index]['labels_task1'].tolist()[0][j] == 'YES':
					Dict_CW[individual][0] = Dict_CW[individual][0] + 1
				else:
					Dict_CW[individual][1] = Dict_CW[individual][1] + 1
			else:
				if sexist_es.loc[sexist_es.id_EXIST==index]['labels_task1'].tolist()[0][j] == 'YES':
					Dict_CW[individual] = [1, 0]
				else:
					Dict_CW[individual] = [0, 1]
	elif index in non_sexist_es.index:
		for (j, individual) in enumerate(ans):
			if Dict_CW.get(individual) is not None:
				if non_sexist_es.loc[non_sexist_es.id_EXIST==index]['labels_task1'].tolist()[0][j] == 'NO':
					Dict_CW[individual][0] = Dict_CW[individual][0] + 1
				else:
					Dict_CW[individual][1] = Dict_CW[individual][1] + 1
			else:
				if non_sexist_es.loc[non_sexist_es.id_EXIST==index]['labels_task1'].tolist()[0][j] == 'NO':
					Dict_CW[individual] = [1, 0]
				else:
					Dict_CW[individual] = [0, 1]
```

```python
ann_keys = list(Dict_CW.keys())
ann_keys_int = [int(s.split('_')[1]) for s in ann_keys]
ann_keys_new = [x for _,x in sorted(zip(ann_keys_int, ann_keys))]
Dict_CW_new = {i: Dict_CW[i] for i in ann_keys_new}
Dict_CW_new
```

```python
Dict_CW_list = list(Dict_CW_new.items())
import numpy as np
Trust_Score = {i[0]: i[1][0]/np.sum(np.array(i[1])) for i in Dict_CW_list}
Trust_Score
```

```python
annotators_column = df.annotators.to_numpy().flatten().tolist()
age_column = df.age_annotators.to_numpy().flatten().tolist()
Dict_ann_age = {}
for (i, ans) in enumerate(annotators_column):
    for (j, individual) in enumerate(ans):
        if Dict_ann_age.get(age_column[i][j]) is None:
            Dict_ann_age[age_column[i][j]] = [individual]
        else:
            if individual not in Dict_ann_age[age_column[i][j]]:
                Dict_ann_age[age_column[i][j]].append(individual)

Dict_ann_age
```

```python
ages = list(Dict_ann_age.keys())
Dict_age_trust = {}
for age in ages:
    sum_trust = np.sum(np.array([Trust_Score[i] for i in Dict_ann_age[age]]))
    Dict_age_trust[age] = sum_trust/len(Dict_ann_age[age])

Dict_age_trust
```

```python
ambiguous_en = ambiguous.loc[ambiguous.lang == 'en']
ambiguous_es = ambiguous.loc[ambiguous.lang == 'es']
ambiguous_en_yes = []
ambiguous_en_no = []
ambiguous_en_neutral = []
for i in range(len(ambiguous_en)):
    yes_score = 0
    no_score = 0
    for j in range(0, len(ambiguous_en.annotators.iloc[i])):
        if ambiguous_en.annotators.iloc[i][j] in Trust_Score:
            if ambiguous_en.labels_task1.iloc[i][j] == 'YES':
                yes_score += Trust_Score[ambiguous_en.annotators.iloc[i][j]]
            else:
                no_score += Trust_Score[ambiguous_en.annotators.iloc[i][j]]
    if yes_score > no_score:
        ambiguous_en_yes.append(ambiguous_en.tweet.iloc[i])
    elif yes_score < no_score:
        ambiguous_en_no.append(ambiguous_en.tweet.iloc[i])
    else:
        ambiguous_en_neutral.append(ambiguous_en.tweet.iloc[i])

print(len(ambiguous_en_yes))
print(len(ambiguous_en_no))
print(len(ambiguous_en_neutral))
```

```python
annotators_column = df.annotators.to_numpy().flatten().tolist()
gender_column = df.gender_annotators.to_numpy().flatten().tolist()
Dict_ann_gender = {}
for (i, ans) in enumerate(annotators_column):
    for (j, individual) in enumerate(ans):
        if Dict_ann_gender.get(gender_column[i][j]) is None:
            Dict_ann_gender[gender_column[i][j]] = [individual]
        else:
            if individual not in Dict_ann_gender[gender_column[i][j]]:
                Dict_ann_gender[gender_column[i][j]].append(individual)

Dict_ann_gender
```

```python
genders = list(Dict_ann_gender.keys())
Dict_gender_trust = {}
for gender in genders:
    sum_trust = np.sum(np.array([Trust_Score[i] for i in Dict_ann_gender[gender]]))
    Dict_gender_trust[gender] = sum_trust/len(Dict_ann_gender[gender])

Dict_gender_trust
```

```python
Dict_ann_age_and_gender = {}
for (i, ans) in enumerate(annotators_column):
    for (j, individual) in enumerate(ans):
        age_and_gender = age_column[i][j] + gender_column[i][j]
        if Dict_ann_age_and_gender.get(age_and_gender) is None:
            Dict_ann_age_and_gender[age_and_gender] = [individual]
        else:
            if individual not in Dict_ann_age_and_gender[age_and_gender]:
                Dict_ann_age_and_gender[age_and_gender].append(individual)

Dict_ann_age_and_gender
        
```

```python
age_and_genders = list(Dict_ann_age_and_gender.keys())
Dict_age_and_gender_trust = {}
for age_and_gender in age_and_genders:
    sum_trust = np.sum(np.array([Trust_Score[i] for i in Dict_ann_age_and_gender[age_and_gender]]))
    Dict_age_and_gender_trust[age_and_gender] = sum_trust/len(Dict_ann_age_and_gender[age_and_gender])

Dict_age_and_gender_trust
```

```python
def create_label(row):
    labels=row['labels_task1']
    annotators=row['annotators']
    ages=row['age_annotators']
    genders=row['gender_annotators']
    Yes_score=0;
    No_score=0;
    count=0
    for i in range(6):
        z=ages[i]+genders[i]
        if labels[i]=='YES':
            Yes_score=Yes_score+Dict_age_and_gender_trust[z]
        if labels[i]=='NO':
            No_score=No_score+Dict_age_and_gender_trust[z]
    if Yes_score>No_score:
        return 'YES'
    if Yes_score<No_score:
        return 'NO'
    if Yes_score==No_score:
        return pd.NaT
```

```python
import preprocessor as p
```

```python
def preprocess_tweet(row):
    text = row['tweet']
    text = p.clean(text)
    return text
```

```python
train['tweet'] = train.apply(preprocess_tweet, axis=1)
dev['tweet'] = dev.apply(preprocess_tweet, axis=1)
```

```python
from gensim.parsing.preprocessing import remove_stopwords
```

```python
def stopword_removal(row):
    text = row['tweet']
    text = remove_stopwords(text)
    
    return text
```

```python
dev_y = dev.iloc[:, -1].values
dev_y =pd.Series(dev_y).map({'YES': 1, 'NO': 0})
y = train.iloc[:, -1].values
y =pd.Series(y).map({'YES': 1, 'NO': 0})
dev_x = dev.iloc[:, 2].values
```

```python
import re
train['tweet'] = train.apply(stopword_removal, axis=1)
train['tweet'] = train['tweet'].str.lower()
train.tweet = train.tweet.apply(lambda x: re.sub(r'https?:\/\/\S+', '', x))
train.tweet.apply(lambda x: re.sub(r"www\.[a-z]?\.?(com)+|[a-z]+\.(com)", '', x))
train.tweet = train.tweet.apply(lambda x: re.sub(r'&[a-z]+;', '', x))
train.tweet = train.tweet.apply(lambda x: re.sub(r"[^a-z\s\(\-:\)\\\/\];='#]", '', x))
train.tweet = train.tweet.apply(lambda x: re.sub(r'@mention', '', x))
dev['tweet'] = dev.apply(stopword_removal, axis=1)
dev['tweet'] = dev['tweet'].str.lower()
dev.tweet = dev.tweet.apply(lambda x: re.sub(r'https?:\/\/\S+', '', x))
dev.tweet.apply(lambda x: re.sub(r"www\.[a-z]?\.?(com)+|[a-z]+\.(com)", '', x))
dev.tweet = dev.tweet.apply(lambda x: re.sub(r'&[a-z]+;', '', x))
dev.tweet = dev.tweet.apply(lambda x: re.sub(r"[^a-z\s\(\-:\)\\\/\];='#]", '', x))
dev.tweet = dev.tweet.apply(lambda x: re.sub(r'@mention', '', x))
```

```python
corpus =[]
dev_corpus=[]
for tweet in train.tweet:
	corpus.append(tweet)
for tweet in dev.tweet:
	dev_corpus.append(tweet)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000)
X=vectorizer.fit_transform(corpus)
dev_x=vectorizer.transform(dev_corpus)
```

```python
X_train=X
y_train=y
X_test=dev_x
y_test=dev_y
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report
classification_report = classification_report(y_test, y_pred)

print('\n Accuracy: ', accuracy_score(y_test, y_pred))
print('\nClassification Report')
print('======================================================')
print('\n', classification_report)
```
